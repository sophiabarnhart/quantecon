## LAKE MODEL
 
Consider a lake model for workers' location. Workers can live in New York or Chicago. The transition probability for different groups of workers are given by
 
$\beta$:  transition probability from New York to Chicago
 
$\gamma$:   transition probability from Chicago to New York
 
 
Let $N_t$ denote the total amount of workers live in New York, $C_t$ denote the total amount of   workers live in Chicago, and $P_t=N_t+C_t$ denote the total amount of labor force. Assume there is no born or death. 

### Question 1 ###

*Given $N_t$ and $C_t$, write down the law of motion for $N_{t+1}$, $C_{t+1}$. Also write down the law of motion for $P_{t+1}$.*

**Law of motion for $N_t$:**
$$ N_{t+1} = N_t \left(\begin{matrix}(1-\beta)\\\gamma\end{matrix}\right)  $$

$$ N_{t+1} = (1-\beta) N_t + \gamma C_t $$

Of the mass of workers $N_t$ who are in New York at time t:
* $(1-\beta) N_t$ will stay in New York at time t+1
* $\gamma N_t$ will move from Chicago to New York at time t+1


**Law of motion for $C_t$:**
$$ C_{t+1} = C_t \left(\begin{matrix}\beta\\(1-\gamma)\end{matrix}\right)  $$

$$ C_{t+1} = \beta N_t + (1-\gamma) C_t $$

Of the mass of workers $C_t$ who are in Chicago at time t:
* $(1-\gamma) C_t$ will stay in Chicago at time t+1
* $\beta C_t$ will move from New York to Chicago at time t+1

**Law of motion for $P_t$:**

\begin{align}
P_t &= N_t + C_t \\
P_{t+1} &= (1+b-d) P_t \\
&= (1-g)P_t
\end{align}

No born (b) or death (d), so **g = 0** (no growth)

Therefore, $$ P_{t+1} = P_t $$

### Question 2 ###

*Find the $A$ matrix such that*

$$ \left(\begin{matrix}N_{t+1}\\C_{t+1}\end{matrix}\right) =A  \left(\begin{matrix}N_t\\C_t\end{matrix}\right)  $$

Linear state space

$$ X_t = \left(\begin{matrix}N_t\\C_t \end{matrix}\right) $$

**Law of motion for $X$:**

$$ X_{t+1} = \left(\begin{matrix}N_{t+1}\\C_{t+1}\end{matrix}\right) =A  \left(\begin{matrix}N_t\\C_t\end{matrix}\right) = A X_t $$

where

$$ A=\left(\begin{matrix}(1-\beta)&\gamma\\\beta&(1-\gamma)\end{matrix}\right)   $$




### Question 3 ###

*Define $n_t=\frac{N_t}{P_t}$ as the fraction of workers in New York, and $c_t=\frac{C_t}{P_t}$ as the fraction of workers in Chicago. Write down the matrix $\widehat A$ such that*

$$  \left(\begin{matrix}n_{t+1}\\c_{t+1}\end{matrix}\right) =\widehat A  \left(\begin{matrix}n_t\\c_t\end{matrix}\right) $$

$$ =\ A \left(\begin{matrix}N_t / P_t\\C_t / P_t\end{matrix}\right) $$

$$ =\ A \left(\begin{matrix}N_t\\C_t\end{matrix}\right) \frac{1}{P_{t+1}}$$

$$ \frac{1}{P_{t+1}} \left(\begin{matrix}N_{t+1}\\C_{t+1}\end{matrix}\right) =\ A \left(\begin{matrix}N_t\\C_t\end{matrix}\right) \frac{1}{P_{t+1}} $$

$P_{t+1}$ = $(1+g)P_t$

$$ =\ A \left(\begin{matrix}N_t\\C_t\end{matrix}\right) \frac{1}{P_t} \frac{1}{1+g} $$

$$ \left(\begin{matrix}\frac{N_{t+1}}{P_{t+1}}\\\frac{C_{t+1}}{P_{t+1}}\end{matrix}\right) =\ \frac{A}{1+g} \left(\begin{matrix}N_t / P_t\\C_t / P_t\end{matrix}\right) $$

$$ \left(\begin{matrix}n_{t+1}\\c_{t+1}\end{matrix}\right) =\ \frac{A}{1+g}  \left(\begin{matrix}n_t\\c_t\end{matrix}\right)$$

Therefore $$ \widehat A =\ \frac{A}{1+g} $$

$$ =\left(\begin{matrix}(1-\beta)/(1+g)&\beta/(1+g)\\\gamma/(1+g)&(1-\gamma)/(1+g)\end{matrix}\right)$$

$X_{t+1} = \widehat A X_t$

Because g=0, ** $\widehat A$ = A**, therefore $X_{t+1}$ = $X_t$

### Question 4 ###

*Find the steady state  $n$ and  $c$ such that*

$$ \left(\begin{matrix}n\\c \end{matrix}\right) =\widehat A  \left(\begin{matrix}n\\c \end{matrix}\right) $$

Components must satisfy

$$ \bar{n} + \bar{c} = 1 $$

Remember

$$ \widehat A = A $$

In steady state, $\widehat A$ is associated with a **unit eigenvalue**

$$ \left(\begin{matrix}n\\c \end{matrix}\right) = \left(\begin{matrix} 1 - \beta & \gamma \\\beta & 1 - \gamma \end{matrix}\right) \left(\begin{matrix} n \\ c \end{matrix}\right)$$

$$\left(\begin{matrix} 1 - \beta - \lambda & \gamma \\ \beta & 1 - \gamma - \lambda \end{matrix}\right)$$
$$\left(\begin{matrix} \beta & -\gamma & 0 \\ 0 & 0 & 0 \end{matrix}\right) \left(\begin{matrix} x_1 \\ x_2 \end{matrix}\right)$$
Multiplying the matrices out, we get
$$x_1 = \frac{\gamma}{\beta}$$
$$ x = \left(\begin{matrix} x_1 \\ x_2 \end{matrix} \right) = \left(\begin{matrix} \frac{\gamma}{\beta} \\ 1 \end{matrix}\right)x_2 = \left(\begin{matrix} \gamma \\ \beta \end{matrix} \right) x_2$$

Therefore

$$ \widehat A = \left(\begin{matrix}\frac{\gamma}{(\gamma + \beta)} \\ \frac{\beta}{(\gamma + \beta)} \end{matrix} \right)$$

### Question 5 ###

*Suppose a worker is at Chicago at time $t$. What is the probability the worker is  New York at time $t+3$?  *

4 possible scenarios:

1. CCCN = $(1-\gamma)(1-\gamma)\gamma$

Therefore

### Question 6 ###

*Simulate the $c_t$ and $n_t$ for 1000 periods. Assume $n_t=0.5$ and $c_0=0.5$. Also assume $\lambda=0.3$ and $\alpha=0.15$. Confirm that $c_t$ and $n_t$ converge to the steady state you have solved earlier. *

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

  
"""
Lake Model
p_t+1 = A p_t

p = state variable containing [n_t, c_0]

beta:  transition probability from New York to Chicago
gamma:   transition probability from Chicago to New York

What is the difference between lambda & alpha vs. beta & gamma?
""" 

# Parameters #

beta = 0.3
gamma  = 0.15
#g = 0 no born or death

n_t = 0.5
c_0 = 0.5

N = 1000

# Transition matrix #

A = np.array(([1-beta,gamma],[beta,1-gamma]))
 
# Setting up simulation #
  
x_rate = np.zeros((N,2))

x_rate[0,:]= np.array(([n_t,c_0]))

# Simulation #

for i in range(1,N):
    x_rate[i,:] =  np.dot(A, x_rate[i-1,:])


# Plot #
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x_rate, alpha = 1) 

ax.set_xlabel(r'time', fontsize=16)
ax.set_ylabel(r'value', fontsize=16)
plt.title(r"Convergence of c_0 and n_t to steady state", fontsize=16, fontweight='bold')
plt.show()
```

The plot shows that yes, $n_t$ and $c_t$ converge at the steady state ($n_t$ = 2/3, $c_t$ = 1/3).

1. each element is non-negative
2. each row sums to one

(2) *Check whether the following matrices are irreducible and aperiodic. *

### Conditions for irreducibility: ###

It is possible to reach every state from all states via finite steps (**all steps communicate**)

### Conditions for aperiodicity: ###

If Markov chain cycles in predictible way, it is **aperiodic** (otherwise, *periodic*)

---

$$ P_1=\left(\begin{matrix}0&0&1\\0&0&1\\\frac{1}{3}&\frac{1}{3}&\frac{1}{3}\end{matrix}\right)   $$

All states can reach each other, so $P_1$ is **IRREDUCIBLE**

Number of periods to return to original state:
* **State 1**: 2  3  4  5  ...
* **State 2**: 2  3  4  5  ...
* **State 3**: 1  2  3  4  ...

Periods have LCD of 1, so $P_1$ is **APERIODIC**

$$ P_2=\left(\begin{matrix}0&1&0\\\frac{1}{2}&0&\frac{1}{2}\\\frac{1}{4}&\frac{1}{4}&\frac{1}{2}\end{matrix}\right)   $$

Periods have LCD of 1, so $P_4$ is **APERIODIC**

### Stochastic Matrix 1

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

"""
Markov Chain
N: number of states
P: Stochastic Matrix (if N=3, should be 3x3 matrix)
T: number of periods for transition

initial state = 1
"""
 
# == Parameters == #

N = 3
T = 1000

P1 = np.array(([0,0,1],[0,0,1],[1/3,1/3,1/3]))

"""
P2 = np.array(([0,1,0],[1/2,0,1/2],[1/4,1/4,1/2]))

P3 = np.array(([0,1,0],[1/2,1/2,0],[1/4,1/4,1/2]))

P4 = np.array(([0,1/2,1/2],[1/2,1/2,0],[1/4,1/4,1/2]))"""


#[state 1],[state 2],[state3]
 
# == Set up simulation == #
  
X  = np.zeros((T), dtype=np.int)

# X: sequence of states (from 1-3, integer)
    
X[0]=0

# initial point

Shock = np.random.uniform(0, 1, T)

# == Simulation == #

for i in range(1,T): #iterating through each period
    X[i]=0      
    for j in range(1,N):           
        if  Shock[i]>sum(P1[X[i-1],0:j]) and  Shock[i]<=sum(P1[X[i-1],0:j+1]):                             
            X[i]=j 
            
           
# == Plots == #
     
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, T)
ax.set_xlabel(r'time', fontsize=16)
ax.set_ylabel(r'state', fontsize=16)   
ax.plot(X, 'bo--', lw=2, alpha=0.8, label='Path')   
ax.legend(ncol=1)
plt.title(r"Path for Stochastic Matrix 1", fontsize=14, fontweight='bold')   


    
plt.show()   

#0.0 = state 1, 1.0 = state 2, 2.0 = state 3
```
